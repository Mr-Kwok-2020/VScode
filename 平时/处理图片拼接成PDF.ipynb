```python
import os
from PIL import Image
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
import tempfile

# 定义文件夹A的路径
folder_A = r"E:\aaaa"

# 获取文件夹A中的所有子文件夹
subfolders = [subfolder for subfolder in os.listdir(folder_A) if os.path.isdir(os.path.join(folder_A, subfolder))]
Cnt = 0
# 遍历每个子文件夹
for subfolder in subfolders:
    Cnt += 1
    # 子文件夹的完整路径
    subfolder_path = os.path.join(folder_A, subfolder)
    
    # 定义文件夹路径和输出PDF文件名
    input_folder = os.path.join(folder_A, subfolder)
    print(input_folder)
    output_pdf_template = subfolder+"_{}.pdf"  # 使用占位符来标明PDF文件名范围

    # 获取文件夹中的所有图片文件
    image_files = [f for f in os.listdir(input_folder) if f.endswith(".jpg") or f.endswith(".png")]

    # 按照文件名排序图片文件列表
    image_files.sort()

    # 设置每个PDF中的图片数量上限
    images_per_pdf = 1000

    # 初始化计数器
    image_count = 0
    pdf_count = 1

    # 创建第一个PDF
    output_pdf = output_pdf_template.format(pdf_count)
    c = canvas.Canvas(output_pdf, pagesize=letter)

    # 获取第一张图片的宽度作为目标宽度
    first_image_path = os.path.join(input_folder, image_files[0])
    target_width = Image.open(first_image_path).width

    # 设置PDF页面尺寸为目标宽度和图片的高度
    for image_file in image_files:
        if image_count == images_per_pdf:
            # 达到图片数量上限，保存当前PDF并创建新的PDF
            c.showPage()
            c.save()

            # 更新计数器和PDF文件名
            pdf_count += 1
            output_pdf = output_pdf_template.format(pdf_count)
            print(output_pdf)
            c = canvas.Canvas(output_pdf, pagesize=letter)
            image_count = 0

        image_path = os.path.join(input_folder, image_file)
        img = Image.open(image_path)

        # 使用指定的重采样滤波器（例如BILINEAR）来调整图片大小
        img = img.resize((target_width, int(img.height * (target_width / img.width))), Image.BILINEAR)

        # 创建一个临时文件来保存调整大小后的图片
        with tempfile.NamedTemporaryFile(delete=False, suffix=".png") as temp_image:
            img.save(temp_image, format="PNG")
            temp_image_path = temp_image.name

        c.setPageSize((target_width, img.height))
        c.drawImage(temp_image_path, 0, 0)
        c.showPage()

        image_count += 1

    # 保存最后一个PDF
    c.showPage()
    c.save()
    print('当前已经处理了',Cnt)
    # if Cnt==25:
    #     break
```

##### 要遍历文件夹A中的所有子文件夹的名称，删除包含数字、'END'、'-'的部分，并保留剩下的部分，你可以使用正则表达式来处理文件夹名称。

```python
import os
import re


# 定义要遍历的文件夹A的路径
folder_A = r"E:\aaaa"

# 遍历文件夹A中的所有子文件夹
for root, dirs, files in os.walk(folder_A):
    for folder_name in dirs:
        folder_path = os.path.join(root, folder_name)

        # 使用正则表达式替换文件夹名称中的数字、'END'、'-'部分
        new_folder_name = re.sub(r'\d+|END||-', '', folder_name)

        # 如果新文件夹名称不为空
        if new_folder_name:
            # 构建新的文件夹路径
            new_folder_path = os.path.join(root, new_folder_name)

            # 重命名文件夹
            os.rename(folder_path, new_folder_path)

print("处理完成")
```

#### 我们遍历源文件夹中的所有子文件夹，然后遍历每个子文件夹中的子文件夹。对于每个子文件夹，我们将其移动到指定的目标文件夹。

```python
import os
import shutil

# 源文件夹路径
source_folder = r"E:\aaaaa2"
# 目标文件夹路径
destination_folder = r"E:\aaaa"

# 遍历源文件夹中的所有子文件夹
for root, dirs, files in os.walk(source_folder):
    for dir in dirs:
        dir_path = os.path.join(root, dir)

        # 遍历子文件夹中的子文件夹
        for subdir in os.listdir(dir_path):
            subdir_path = os.path.join(dir_path, subdir)

            # 将子文件夹移动到目标文件夹
            new_subdir_path = os.path.join(destination_folder, subdir)
            shutil.move(subdir_path, new_subdir_path)
            print(f"移动子文件夹 '{subdir}' 到目标文件夹")

print("处理完成")
```

#### 在解压子文件夹后，遍历每个子文件夹，然后将子文件夹的名字改成压缩包的名字

```python
import os
import zipfile

# 定义要解压的文件夹路径
folder_to_unzip = r'E:\aaaaa2'

# 获取文件夹中的所有压缩文件
zip_files = [f for f in os.listdir(folder_to_unzip) if f.endswith(".zip")]

# 遍历每个ZIP文件并解压到以压缩包名命名的文件夹
for zip_file in zip_files:
    zip_file_path = os.path.join(folder_to_unzip, zip_file)
    
    # 提取压缩包的名字作为解压文件夹的名字
    output_folder_name = os.path.splitext(zip_file)[0]
    output_folder_path = os.path.join(folder_to_unzip, output_folder_name)

    # 创建输出文件夹
    os.makedirs(output_folder_path, exist_ok=True)

    # 打开ZIP文件并解压到输出文件夹
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        zip_ref.extractall(output_folder_path)

    print(f"解压 {zip_file} 到 {output_folder_name} 完成")

    # 遍历子文件夹并重命名
    for root, dirs, files in os.walk(output_folder_path):
        for dir in dirs:
            dir_path = os.path.join(root, dir)
            new_dir_name = output_folder_name
            os.rename(dir_path, os.path.join(root, new_dir_name))

print("所有文件解压和重命名完成")



```
